```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
sns.scatterplot(data = df,  x = "cgpa", y = "resume_score", hue = df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```
