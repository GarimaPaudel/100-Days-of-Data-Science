**Removing Null values**

```python
# Dropping rows with null values
df1 = df.dropna() #by default, axis = 0
```

```python
df1.shape
```

```python
df1.head()
```

Initiall we had (891, 12) shaped data. After dropping all the rows with null values, it resulted the dataset(stored in df1) of shape (183,12).
i.e 708 rows got deleted.

```python
#Dropping columns with null values
df2 = df.dropna(axis = 1) #axis = 0 for rows and axis = 1 for columns
```

```python
df2.shape
```

Initially we had (891, 12) shaped data. After dropping all the columns with null values, it resulted the dataset(stored in df2) of shape (891, 9).
i.e 3 columns (Age, Embarked, Cabin) got deleted.

```python
df2.head()
```

**Filling Null values**

```python
df.head()
```

```python
df.fillna(0) #fills null values with 0, we can give other values as well
```

```python
df.fillna({'Age':'none','Cabin':0})  # we can even fill particular column's null values like this
```

**Forward Fill and Backward Fill**
- For time-series data, you might use methods like forward-fill or backward-fill to propagate the last known value forward or backward in time.

```python
df.fillna(method="ffill") # ffill fills the null values with previous/preceeding value
```

```python
df.fillna(method = 'ffill', axis=1) # null values is filled with the previous columns value 
```

```python
# we can see the age value of 888 index row is filled with female, we cannot fill age data with values of Sex column.
#So, we should be careful to check which way we want to fill the null values with.
```

```python
df.fillna(method="bfill") #bfill for Backward Fill
```

**Filling null values with Mean value**

```python
df.tail()
```

```python
mean_age = df['Age'].mean()
mean_age
```

```python
 df['Age'].fillna(mean_age)
```

```python
# we can see the age of 888th row was null, but now it is filled with average age value
```

**Filling null with mode value**

```python
mode_embarked = df['Embarked'].mode()
mode_embarked
```

```python
df['Embarked'].fillna(mode_embarked)
```

**Interpolation**

```python
# Sample dataset with null values
data = {'Timestamp': ['2023-01-01', '2023-01-02', None, '2023-01-04', None],
        'Value': [10, 15, None, 25, None]}

df6 = pd.DataFrame(data)
df6['Timestamp'] = pd.to_datetime(df6['Timestamp'])

# Sort the DataFrame by timestamp
df6 = df6.sort_values('Timestamp')

# Interpolate missing values using linear interpolation
df6['Value'] = df6['Value'].interpolate()

# Print the dataset after interpolation
print("Dataset after interpolation:")
print(df6)
```
