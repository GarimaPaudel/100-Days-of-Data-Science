```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
data, labels = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)
```

```python
# Create and fit a DBSCAN model
dbscan = DBSCAN(eps=0.6, min_samples=5)
dbscan.fit(data)
```

```python
# Extract the cluster labels
cluster_labels = dbscan.labels_
```

```python
# Number of clusters found (including noise, labeled as -1)
num_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
```

```python
# Plot the data points and their cluster assignments
unique_labels = set(cluster_labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
```

```python
for cluster_label, color in zip(unique_labels, colors):
    if cluster_label == -1:
        color = 'k'  # Assign black color to noise points

    class_member_mask = (cluster_labels == cluster_label)
    xy = data[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=[color], label=f'Cluster {cluster_label}')
    
plt.title(f'DBSCAN Clustering (Found {num_clusters} clusters)')
plt.legend()
plt.show()
```

In this code:

- We generate synthetic data using make_blobs from scikit-learn, creating three clusters with some random noise.

- We create a DBSCAN model with specified parameters (eps and min_samples) and fit it to the data.

- We extract the cluster labels assigned by DBSCAN.

- We count the number of clusters found and plot the data points, coloring them by their cluster assignments. Noise points are assigned the color black.
