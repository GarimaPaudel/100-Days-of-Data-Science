```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(type(X_train))
```

```python
import torch.nn as nn
class CustomerChurn(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(11,64)
        self.l2 = nn.Linear(64,32)
        self.l3 = nn.Linear(32,16)
        self.l4 = nn.Linear(16,8)
        self.l5 = nn.Linear(8,1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.relu(self.l1(x))
        x = self.relu(self.l2(x))
        x = self.relu(self.l3(x))
        x = self.relu(self.l4(x))
        x = self.sigmoid(self.l5(x))
        return x
```

```python
import torch.optim as optim
loss_function = nn.BCELoss()
model = CustomerChurn()
optimizer = optim.Adam(model.parameters(),lr=0.01)
```

```python
y_test = y_test.to_numpy()
y_train = y_train.to_numpy()
```

```python
# Convert to PyTorch tensors
import torch
# Convert X_train and X_test to tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)

# Convert y_train and y_test to tensors, and add an extra dimension using unsqueeze
y_train = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
y_test = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)

 
```

```python
epochs = 100
for epoch in range(epochs):
    model.train()    
      
   # Forward pass
    outputs = model(X_train)
    
    loss = loss_function(outputs, y_train)
    
    optimizer.zero_grad()  
    
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
    
    
        
```

```python
# def accuracy_fn(y_true, y_pred):
#     correct = torch.eq(y_true, y_pred).sum().item()
#     acc = (correct / len(y_pred)) * 100
#     return acc
from sklearn.metrics import accuracy_score
```

```python
model.eval()
with torch.inference_mode():
    y_pred = model(X_test)
   
    predicted = (y_pred >= 0.5).float()
#     print(predicted)
    
    acc = accuracy_score(y_test, predicted)
    print(acc)
    
```
